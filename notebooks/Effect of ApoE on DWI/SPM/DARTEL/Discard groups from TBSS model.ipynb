```python
from glob import glob
import os.path as osp
import string
fp = '/home/grg/tbss_apoe34_vs_apoe33/origdata/'
subjects = [string.atoi(osp.basename(e).split('_')[1][:5]) for e in sorted(glob(osp.join(fp, '*.nii.gz')))]
files = [osp.basename(e) for e in sorted(glob(osp.join(fp, '*.nii.gz')))]
#lines = [e.rstrip('\n') for e in open(fp).readlines()]
import pandas as pd
cov = pd.read_excel('/home/grg/spm/data/covariates.xls')
cov = cov.set_index('subject')

data = []
for s,f in zip(subjects, files):
    row = [s,f]
    d = cov.loc[s].to_dict()
    row.append(d['age'])
    row.append(d['gender'])
    row.append(d['educyears'])
    row.append(int(d['apo']==2.0))
    row.append(int(d['apo']==3.0))
    data.append(row)

df = pd.DataFrame(data)
df.head()
#df.to_excel('/tmp/test.xls')
#df.to_csv('/tmp/test.csv', sep=' ', index=False)
```
