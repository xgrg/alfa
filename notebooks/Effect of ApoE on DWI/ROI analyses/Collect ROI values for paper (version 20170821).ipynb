```python
%run /home/grg/git/alfa/nilearn-helper.py
%matplotlib inline
from glob import glob
import os.path as osp
from nistats import thresholding
import numpy as np
from nilearn import image, plotting
import nibabel as nib
import json
import os
```

### MD recessive

```python
roivalues_wd = '/tmp/roistats_bsc_20170821/MD' # Folder where the files containing the ROI values will be stored
data_wd = '/home/grg/tbss_bsc_subjects/skel/MD/' # Folder containing the images over which the ROI values will be extracted

#subjects = json.load(open(osp.join('/home/grg/spm', 'data', 'subjects.json'))) # List of subjects
subjects = [e.split('/')[-1].split('_')[0] for e in sorted(glob(osp.join('/home/grg/bsc/bsc_ALFA_20170821/data/origdata/*.nii.gz')))]

rois_fp = '/tmp/roistats_bsc_20170821/tbss_MD_model4_recessive.nii.gz'
# Load the collection of ROIs
rois = np.asarray(nib.load(rois_fp).dataobj)
nb_roi = len(np.unique(rois)) - 1
print nb_roi, 'regions - ', len(subjects), 'subjects'

# Iterate over subjects
for i,s in enumerate(subjects):
    try:        
        mdfp = sorted(glob(osp.join(data_wd, 'vol*.nii.gz')))[i]
        
        # Build the command and run it
        cmd = 'AimsRoiFeatures -i %s -s %s -o %s'%(rois_fp, mdfp, osp.join(roivalues_wd, '%s_stats_recessive.csv'%s))
        print cmd
        os.system(cmd)
    except Exception as e:
        print s, e
```
