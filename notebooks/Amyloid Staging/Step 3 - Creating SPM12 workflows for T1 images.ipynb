# Run SPM12 over a set of images using Nipype

This pipeline actually includes two prior coregistration steps (using ICBM and MNI atlases) to help the process and minimize risks of segmentation failures.

The `create_workflow` function takes a list of paths to the input T1 images, a list of associated subject identifiers and a folder where the jobs will be written and run.
The subjects are used to name Nipype nodes and folders after, which makes the identification of results easier. Nevertheless, this could be changed so that folders would be just named simply with integers, hence removing the need for these subject names.

```python
# Database folder
dd = '/home/grg/data/AmylStaging/'

# T1 images to run SPM12 on
sources = glob(osp.join(dd, '*', 'T1', '*_mabonlm_nobias.nii'))[:10]
# Corresponding subjects
subjects = [spm12._get_subject_(e) for e in sources]

w = spm12.create_workflow(sources, subjects, basedir='/tmp/test_spm12')
w.run('MultiProc', plugin_args={'n_procs': 8})
```
